### Parser les plats d'une page

```python
URL_MAIN = "https://www.marmiton.org/recettes/index/categorie/plat-principal/"
driver.get(URL_MAIN)
```

```python
def recette_par_page(driver):
    recettes = []
    # Rechercher les éléments avec la classe 'mrtn-card__title' uniquement à l'intérieur de 'recette'
    titre_elements = driver.find_elements(By.XPATH, './/*[contains(@class, "mrtn-card__title")]/a')
    for titre_element in titre_elements:
        lien_recette = titre_element.get_attribute('href')
        titre_recette = titre_element.text.strip()
        recettes.append((titre_recette, lien_recette))
    
    return recettes
```

### Recuperer une liste de liens pointant vers les recettes

```python
URL_ROOT = "https://www.marmiton.org/recettes/index/categorie/plat-principal"

titre_recette = []
lien_recette = []

for i in range(1, 510):
    URL_PAGE_PATH = f"{URL_ROOT}/{i}"
    driver.get(URL_PAGE_PATH)
    # Rechercher les éléments avec la classe 'mrtn-card__title' uniquement à l'intérieur de 'recette'
    titre_elements = driver.find_elements(By.XPATH, './/*[contains(@class, "mrtn-card__title")]/a')
    for titre_element in titre_elements:
        lien_recette.append(titre_element.get_attribute('href'))
        titre_recette.append(titre_element.text.strip())
```

```python
dataset_lien_recettes = pd.DataFrame({
    'Recette' : titre_recette,
    'Lien recette' : lien_recette
})

dataset_lien_recettes.to_csv('liens-recette.csv', index=False, encoding='utf-8-sig')
```
